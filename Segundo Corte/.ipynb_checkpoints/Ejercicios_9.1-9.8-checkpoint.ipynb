```python
#9.2

A=[[4,7],[1,2],[5,6]]
B=[[4,3,7],[1,2,7],[1,0,4]]
C=[[3],[6],[1]]
D=[[9,4,3,-6],[2,-1,7,5]]
E=[[1,5,8],[7,2,3],[4,0,6]]
F=[[3,0,1],[1,7,3]]
G=[[7,6,4]]


#print (len(A))#Numero de filas 
#print (len(A[0]))#Numero de Columnas
print ('a)')
print ('A[',len(A),',',len(A[0]),']')
print ('B[',len(B),',',len(B[0]),']')
print ('C[',len(C),',',len(C[0]),']')
print ('D[',len(D),',',len(D[0]),']')
print ('E[',len(E),',',len(E[0]),']')
print ('F[',len(F),',',len(F[0]),']')
print ('G[',len(G),',',len(G[0]),']')


print ('\nb)')
M=[A,B,C,D,E,F,G]
abc=['A','B','C','D','E','F','G']

for i in range (0,7):
    p=M[i]
    if len(p)==len(p[0]):
        print (abc[i],' = Cuadrada')
    if len(p[0])==1:
        print (abc[i],' = Columna')
    if len(p)==1:
        print (abc[i],' = Renglon')

        
print ('\nc)')
print ("""A[1][2]=7
B[2][3]=7
D[3][2]= Posicion inexistente en la matriz D
E[2][2]=2
F[1][2]=0
G[1][2]=6
    
""")

print ('\nd)\n')

def uM(a,b):
    matriz = []
    for i in range(a):
        matriz.append([])
        for j in range(b):
            matriz[i].append(None)
    return (matriz)        

def opr(k,a,b):    
    if k<3:
        if(len(a)==len(b)&len(a[0])==len(b[0])):
            u = uM(len(a),len(a[0]))               
            for i in range(len(a)):
                for j in range(0,len(a[0])):
                    if k==1:
                        u[i][j]=a[i][j]+b[i][j]
                    if k==2:    
                        u[i][j]=a[i][j]-b[i][j]                
            print (u)
        else:
            print ('Esta operacion no se puede hacer')
    if k==3:                                 #Matriz por escalar
        u = uM(len(b),len(b[0]))  
        for i in range(len(b)):
            for j in range(0,len(b[0])):
                u[i][j]=7*b[i][j]
        print (u)
    if k==4:                                 #Multiplicacion matricial
        if len(a[0])==len(b):
            u = uM(len(a),len(b[0]))        
            n=0
            for i in range(len(a)):
                for j in range(len(b[0])):
                    for y in range(len(a[0])):
                        n+=a[i][y]*b[y][j]               
                    u[i][j]=n
                    n=0            
            print (u)
        else:
            print('NO es posible efectuar la multiplicacion no son de la forma A[a][b] , B[b][c]')
    if k==5:                                 #Transpuesta
        u = uM(len(b[0]),len(b))                
        for i in range(len(b[0])):
            for j in range(len(b)):
                u[i][j]=b[j][i]                
        return (u)
    if k==6:                                 #Matriz identidad por otra  I*X
        u = uM(len(b[0]),len(b))                
        for i in range(len(b[0])):
            for j in range(len(b)):
                if i==j:
                    u[i][j]=1
                else:
                    u[i][j]=0
        opr(4,u,b)
        
        
    
print('1.E+B:')
opr(1,E,B)
print('\n2.A+F:')
opr(1,A,F)
print('\n3.B-E:')
opr(2,B,E)
print('\n4.7*B:')
opr(3,7,B)
print('\n5.E*B:')
opr(4,E,B)
print('\n6.C(Transpuesta):')
print(opr(5,0,C))
print('\n7.B*A:')
opr(4,B,A)
print('\n8.D(Transpuesta):')
print(opr(5,0,D))
print('\n9.A*C:')
opr(4,A,C)
print('\n10I*B:')
opr(6,0,B)
print('\n11.E transpuesta*E:')
opr(4,opr(5,0,E),E)
print('\n12.C transpuesta*C:')
opr(4,opr(5,0,C),C)


```

```python
"""
9.3

b)Debido a las restricciones propias de las matrices en cuanto a sus dimensiones, 
no se pueden efectuar ciertas multiplicaciones con las matrices dadas.


c)Es importante el orden de la multiplicacion, debido a que esta operación
depende rigurosamente de las dimensiones de las matrices, en donde el numero
de columnas del primer factor deben coincidir con el numero de filas del segundo,
para obtener como resultado una nueva matriz cuyas dimensiones estarán dadas por
las filas del primero y columnas del segundo

"""


A=[[1,6],[3,10],[7,4]]
B=[[1,3],[0.5,2]]
C=[[2,-2],[-3,1]]
M=[A,B,C]
abc=['A','B','C']

def multiplicacion(a,b):
    if len(a[0])==len(b):
        u = uM(len(a),len(b[0]))        
        n=0
        for i in range(len(a)):
            for j in range(len(b[0])):
                for y in range(len(a[0])):
                    n+=a[i][y]*b[y][j]               
                u[i][j]=n
                n=0            
        return (u)
    else:
        return ('No se puede efectuar la multiplicacion')

print('a)\n')    
    
for i in range (len(M)):
    for j in range (len(M)):
        if i!=j:
            print (abc[i],'*',abc[j],':\n ',multiplicacion(M[i],M[j]))
    
        
```

```python
"""
9.6
b)

"""
matriz =[[0,2,5],[2,1,1],[3,1,0]]
matriz2, matriz3, matriz4 =[[0,2,5],[2,1,1],[3,1,0]],[[0,2,5],[2,1,1],[3,1,0]],[[0,2,5],[2,1,1],[3,1,0]]
mresp=[9,9,10]
D,D2,D3,D4=0,0,0,0

def det2x2(m,b):
    r=0
    if b==0:
        r=(m[1][1]*m[2][2]-m[1][2]*m[2][1])
    if b==1:
        r=(m[1][0]*m[2][2]-m[1][2]*m[2][0])*(-1)
    if b==2:
        r=(m[1][0]*m[2][1]-m[1][1]*m[2][0])    
    
    return (r)

for j in range (0,3):
    matriz2[j][0]=mresp[j]
    matriz3[j][1]=mresp[j]
    matriz4[j][2]=mresp[j]

for i in range (0,3):    
    D = D + (matriz[0][i]*det2x2(matriz,i))
    D2 = D2 + (matriz2[0][i]*det2x2(matriz2,i))
    D3 = D3 + (matriz3[0][i]*det2x2(matriz3,i))
    D4 = D4 + (matriz4[0][i]*det2x2(matriz4,i))

    
print ('D = ',D)
print ('X1 = ',(D2/D))
print ("X2 = ",(D3/D))
print ("X3 = ",(D4/D))

"""
c)

"""

print ('2 X2 +5 X3 = 9 ==>',(2*(D3/D))+5*(D4/D))
print ('2 X1 + X2 +X3  = 9 ==>',(2*(D2/D))+(D3/D)+(D4/D))
print ('3 X1 + X2 = 10 ==>',(3*(D2/D))+(D3/D))
```
